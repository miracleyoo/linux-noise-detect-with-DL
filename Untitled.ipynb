```python
def divPrime(num):
    lt = []
    print(num, '=', end=' ')
    while num != 1:
        for i in range(2, int(num+1)):
            if num % i == 0:  # inum
                lt.append(i)
                num = num / i # numi
                break
    for i in range(0, len(lt)-1):
        print(lt[i], '*', end=' ')
 
    print(lt[-1])
```

```python
divPrime(40184+8+256+512)
divPrime(40960)
```
